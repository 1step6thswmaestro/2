```python
from html.parser import HTMLParser as hp
import html
import re
import copy
import numpy as np
import pandas as pd
import json
import glob
from random import shuffle
from bs4 import BeautifulSoup as bs
```

```python
data_path = "data"
#data_path = "data(test)"

file_list = glob.glob("%s/*.json" % data_path)

json_train=[]

shuffle(file_list)
for json_file_name in file_list:
    json_file = json.loads(open(json_file_name).read())
    json_train += json_file["articles"]
```

```python
tag_list = []
label_list = []

num_json_train = len(json_train)

for i, article in enumerate(json_train):
    if i % 10000 == 0:
        print("converting... (%d/%d)" % (i, num_json_train))

    tags = {}
    
    train_text = article["text"]
    soup = bs(train_text, 'html.parser')
    for tag in soup.find_all():
        tag_name = 'TAG_' + tag.name
        if tag_name not in tags:
            tags[tag_name] = 0
        tags[tag_name] += 1

    tags["is_troll"] = article["is_troll"]
    tags["pk"]=article["pk"]
    tag_list.append(tags)
    

data = pd.DataFrame.from_dict(tag_list)
data = data.fillna(0)
data = data.set_index("pk")

data
```

```python
import os
import pickle

pre = data.col

used_model = "tag_counter"
if not os.path.exists(used_model):
    os.makedirs(used_model)

pickle.dump(data[pre], open("%s/tag_counter.p" % (used_model), "wb"), protocol = pickle.HIGHEST_PROTOCOL)
```
